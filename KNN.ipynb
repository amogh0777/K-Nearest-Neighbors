In this project, we use the **Iris dataset**, which includes measurements of three flower species — **Setosa**, **Versicolor**, and **Virginica**. # **Key Terminologies:**

**Overview:**

1. **K-Nearest Neighbors (KNN):**
   A simple supervised machine learning algorithm used for **classification** and **regression**. It predicts the class of a new data point by checking the most common class among its **K nearest neighbors** in the dataset.
2. **Feature:**
   A measurable property or input variable of a data point. In the Iris dataset, features are **sepal length**, **sepal width**, **petal length**, and **petal width**.
3. **Target Variable (Label):**
   The output variable that the model predicts. In this project, it’s the **species of the iris flower**.
4. **Label Encoding:**
   The process of converting **categorical data (like flower names)** into **numeric form** so that the machine learning model can understand it.
5. **Train-Test Split:**
   Dividing the dataset into two parts — one for **training** the model and another for **testing** its performance. Usually, 80% is for training and 20% for testing.
6. **Feature Scaling:**
   Adjusting all features to a similar scale (range) so that no single feature dominates others. It’s especially important for **distance-based algorithms** like KNN.
7. **Accuracy:**
   The percentage of predictions the model got **correct** out of all predictions made.
8. **Confusion Matrix:**
   A table that shows how many predictions were correct and how many were wrong, helping to evaluate the model’s performance.
9. **Classification Report:**
   A summary that shows key metrics such as **precision**, **recall**, and **F1-score** for each class.
10. **Hyperparameter (K):**
    A value set before training that tells the KNN model **how many neighbors** to look at when making a prediction.
